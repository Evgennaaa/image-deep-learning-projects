## Проект: Рекомендательная система одежды на основе изображений

**Цель:** Разработать систему, которая находит похожие товары на основе их визуальных характеристик. Мы будем использовать предобученные сверточные нейронные сети (CNN) для извлечения векторных представлений (эмбеддингов) изображений и сравнивать их с помощью косинусной близости.

### Шаг 3: Подготовка данных и изображений

```python
# Создаем новый столбец с полным путем к файлу изображения
fashion_styles_df['image_filepath'] = fashion_styles_df['id'].apply(lambda img_id: f'images/{img_id}.jpg')

# Проверяем, существует ли каждый файл изображения на диске
fashion_styles_df['is_image_present'] = fashion_styles_df['image_filepath'].apply(os.path.exists)
print(f"Найдено {fashion_styles_df['is_image_present'].sum()} существующих изображений.")

# Отфильтровываем датафрейм, оставляя только те записи, для которых есть изображения
df_with_images = fashion_styles_df[fashion_styles_df['is_image_present']].copy()
print(f"Размер датафрейма после фильтрации: {len(df_with_images)} записей.")
```

```python
def display_sample_images(dataframe, num_items=5):
    """Отображает случайный набор изображений из датафрейма."""
    sample_df = dataframe.sample(num_items)
    plt.figure(figsize=(22, 7))
    for i, (idx, row) in enumerate(sample_df.iterrows()):
        plt.subplot(1, num_items, i + 1)
        try:
            img_obj = Image.open(row['image_filepath'])
            plt.imshow(img_obj)
            # Укорачиваем длинные названия для удобства отображения
            display_name = row['productDisplayName']
            title_text = display_name[:30] + '...' if len(display_name) > 30 else display_name
            plt.title(f"{title_text}\n(ID: {row['id']})")
            plt.axis('off')
        except Exception as e:
            print(f"Не удалось загрузить изображение {row['image_filepath']}: {e}")
    plt.show()

print("\nСлучайные примеры товаров из нашего набора данных:")
display_sample_images(df_with_images)
```

#### Создание рабочей выборки и определение вспомогательных функций

Чтобы ускорить процесс извлечения признаков, мы будем работать с уменьшенной выборкой данных.

```python
# Определяем размер выборки для ускорения вычислений
SAMPLE_SIZE = 1500
data_subset = df_with_images.sample(SAMPLE_SIZE, random_state=42).reset_index(drop=True)

print(f"Для разработки системы будет использоваться выборка из {len(data_subset)} изображений.")

# Стандартизируем размер изображения для всех моделей
TARGET_IMG_SIZE = (224, 224)
```

```python
def process_image_for_nn(filepath, target_dims, preprocess_func):
    """Загружает, изменяет размер и преобразует изображение для подачи в нейросеть."""
    try:
        img = image.load_img(filepath, target_size=target_dims)
        img_array = image.img_to_array(img)
        img_array_expanded = np.expand_dims(img_array, axis=0)
        return preprocess_func(img_array_expanded)
    except Exception as e:
        print(f"Ошибка при обработке файла {filepath}: {e}")
        return None

def extract_features(dataframe, cnn_model, dimensions, preprocess_func):
    """Извлекает эмбеддинги для всех изображений в датафрейме с помощью указанной CNN."""
    feature_list = []
    progress_bar = tqdm(dataframe['image_filepath'], desc=f"Извлечение признаков ({cnn_model.name})")
    for path in progress_bar:
        processed_img = process_image_for_nn(path, dimensions, preprocess_func)
        if processed_img is not None:
            # Получаем вектор признаков и преобразуем его в 1D массив
            feature_vector = cnn_model.predict(processed_img, verbose=0).flatten()
            feature_list.append(feature_vector)
        else:
            # В случае ошибки добавляем нулевой вектор для сохранения структуры данных
            feature_list.append(np.zeros(cnn_model.output_shape[1])) 
    return np.array(feature_list)
```

### Шаг 4: Извлечение признаков (эмбеддингов) с помощью CNN

Мы протестируем три популярные архитектуры: VGG16, ResNet50 и MobileNetV2.

#### Подход 4.1: Модель VGG16

```python
# Загружаем VGG16 с весами ImageNet, без верхних слоев, с пулингом для получения вектора
vgg_model = VGG16(weights='imagenet', include_top=False, pooling='avg', input_shape=(*TARGET_IMG_SIZE, 3))

vgg_features = extract_features(data_subset, vgg_model, TARGET_IMG_SIZE, vgg_preprocess)

print(f"\nРазмерность матрицы признаков VGG16: {vgg_features.shape}")
```

#### Подход 4.2: Модель ResNet50

```python
# Загружаем ResNet50 аналогично VGG16
resnet_model = ResNet50(weights='imagenet', include_top=False, pooling='avg', input_shape=(*TARGET_IMG_SIZE, 3))

resnet_features = extract_features(data_subset, resnet_model, TARGET_IMG_SIZE, resnet_preprocess)

print(f"\nРазмерность матрицы признаков ResNet50: {resnet_features.shape}")
```

#### Подход 4.3: Модель MobileNetV2

```python
# Загружаем легковесную модель MobileNetV2
mobilenet_model = MobileNetV2(weights='imagenet', include_top=False, pooling='avg', input_shape=(*TARGET_IMG_SIZE, 3))

mobilenet_features = extract_features(data_subset, mobilenet_model, TARGET_IMG_SIZE, mobilenet_preprocess)

print(f"\nРазмерность матрицы признаков MobileNetV2: {mobilenet_features.shape}")
```

### Шаг 5: Построение и оценка рекомендательной системы

```python
def get_recommendations(item_index, feature_matrix, top_n=5):
    """Находит и возвращает индексы N самых похожих товаров."""
    # Вектор признаков для исходного товара
    source_vector = feature_matrix[item_index].reshape(1, -1)
    
    # Расчет косинусной близости между исходным и всеми остальными товарами
    similarity_scores = cosine_similarity(source_vector, feature_matrix)
    
    # Сортировка индексов по убыванию значения близости
    sorted_item_indices = np.argsort(similarity_scores[0])[::-1]
    
    # Возвращаем top_n индексов, пропуская первый (это сам исходный товар)
    return sorted_item_indices[1:top_n+1]

def show_recommendations(item_index, feature_matrix, dataframe, model_name):
    """Визуализирует исходный товар и найденные для него рекомендации."""
    
    # Данные по исходному товару
    source_item_data = dataframe.iloc[item_index]
    plt.figure(figsize=(22, 8))
    
    # Отображение исходного товара
    ax = plt.subplot(1, 6, 1)
    ax.imshow(Image.open(source_item_data['image_filepath']))
    ax.set_title(f"Исходный товар\n({model_name})")
    ax.axis('off')
    
    # Поиск и отображение рекомендованных товаров
    recommended_indices = get_recommendations(item_index, feature_matrix, top_n=5)
    
    print(f"--- Результаты для модели: {model_name} ---")
    for i, rec_idx in enumerate(recommended_indices):
        rec_item_data = dataframe.iloc[rec_idx]
        ax = plt.subplot(1, 6, i + 2)
        ax.imshow(Image.open(rec_item_data['image_filepath']))
        ax.set_title(f"Рекомендация №{i+1}")
        ax.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
# Выбираем случайный товар для тестирования системы
test_item_idx = random.randint(0, len(data_subset) - 1)

print(f"Тестируем систему на товаре с индексом {test_item_idx}: {data_subset.iloc[test_item_idx]['productDisplayName']}\n")

# Отображаем рекомендации от каждой из трех моделей
show_recommendations(test_item_idx, vgg_features, data_subset, 'VGG16')
show_recommendations(test_item_idx, resnet_features, data_subset, 'ResNet50')
show_recommendations(test_item_idx, mobilenet_features, data_subset, 'MobileNetV2')
```

### Шаг 6: Сохранение результатов

Сохраним эмбеддинги, полученные с помощью наилучшей модели (предположим, это ResNet50), и соответствующий им срез метаданных для дальнейшего использования, например, в веб-приложении.

```python
# Выбираем эмбеддинги от модели, показавшей лучшие результаты (например, ResNet50)
chosen_features = resnet_features

# Определяем пути для сохранения артефактов
features_output_path = 'resnet_fashion_embeddings.npy'
metadata_output_path = 'fashion_metadata_subset.pkl'

# Сохраняем матрицу с векторными представлениями изображений
np.save(features_output_path, chosen_features)
print(f"Матрица эмбеддингов размером {chosen_features.shape} сохранена в файл: {features_output_path}")

# Сохраняем датафрейм с метаданными для использованной выборки
data_subset.to_pickle(metadata_output_path)
print(f"Срез метаданных ({len(data_subset)} записей) сохранен в файл: {metadata_output_path}")
```
